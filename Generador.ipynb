# Generador de Ondas en Python

## Utilizamos codigo para generar ondas de distinta frecuencia y graficarlas.
## Luego utilizamos funciones de onda predefinidas.

### Onda de 500 Hz

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la onda

vmax = 1 # Amplitud maxima
dc = 0 # Valor medio
f0 = 500 # Hz
ph = 0*np.pi # Fase

fs = 1000 # frecuencia de muestreo
N = 1000 # Cantidad de muestras

ts = 1/fs # tiempo de muestreo

# Grilla de sampleo temporal
tt = np.linspace(0, (N-1)*ts, N)
argg = 2*np.pi*f0*tt+ph

xx = np.sin(argg)*vmax

plt.plot(tt, xx)
```

#### La frecuencia de la onda es igual a la frecuencia de Nyquist, entonces los puntos de muestreo estan muy separados entre si y la reconstrucción no es precisa.

### Onda de 999 Hz

```python
f0 = 999 # Hz

argg = 2*np.pi*f0*tt+ph
xx = np.sin(argg)*vmax
plt.plot(tt, xx)
```

#### La reprecentación de la onda porque la frecuecnia de muestreo y la de la onda original practicamente coinciden. Es decir que cada muetsra se toma en el mismo punto que la sinusoide.

### Onda de 1001 Hz

### Onda de 2001 Hz

```python
f0 = 2001 # Hz

argg = 2*np.pi*f0*tt+ph
xx = np.sin(argg)*vmax
plt.plot(tt, xx)
```

#### La representación de la onda queda desfasada. Al tener el doble de frecuencia de la onda en comparación con la frecuencia de muestreo, las muestras caen en puntos alternados del ciclo. 

### Onda predeterminada - Onda Cuadrada

```python
from scipy import signal

xx = signal.square(tt, duty=0.075)
plt.plot(tt, xx)
```
