Analizamos intuitivamente el comportamiento del circuito en extremos de frecuencia para determinar el comportamiento de la red y definir de que tipo de filtro se trata.
De esta forma obtenemos informacion previa al analisis para saber que características debería presentar el circuito y comparar los resultado obtenidos.

### Circuitos:

En un primer analisis, consideramos las impedancias:

<p style='text-align: center;'>$Zc = 1/sC$ 
<p style='text-align: center;'>$Zl = sL$ 
<p style='text-align: center;'>$Zr = R$

<p style='text-align: left;'>Teniendo en cuenta esto, analizamos cada circuito:
<p style='text-align: left;'>Para el circuito A, cuando las frecuencias tienden a 0, la impedancia del capacitor tiende a infinito, por lo que no hay paso de corriente (se comporta como un circuito abierto). En el extremo opuesto, cuando las frecuencias tienden al infinito, la impedancia de la bobina tiende a infinito, por lo que no hay un paso de corriente tampoco en ese caso. Es decir que el caso intermedio es donde resta analizar, y estariamos hablando entonces de un filtro pasa bandas.
<p style='text-align: left;'>Para el circuito B, cuando las frecuencias tienden a 0, sucede lo mismo que en el caso anterior, por lo que se filtrarán las bajas frecuencias. En el caso de las frecuencias en el infinito, la bobina tiene impedancia infinita y el circuito estaría reducido a un filtro RC convencional. Entonces estaríamos hablando de un filtro pasa altos, ya que no permite el paso de frecuencias bajas.
<p style='text-align: left;'>Segun lo leido en bibliografía, la frecuencia de corte será donde la ganancia sera -3dB. En el caso del filtro pasa banda, habra dos frecuencias (*f1* y *f2*) que determinarán el ancho de banda y que ceran "equidistantes" a la frecuencia de corte.
<p style='text-align: left;'>En el filtro pasa altos, la frecuencia de corte determinará a partir de donde comienza el ancho de banda, que, en principio, sería infinita hacia las frecuencias en el infinito.
