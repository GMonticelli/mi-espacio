## Modelizando un ADC

En este trabajo modelizamos una señal señal senoidal, la cual contaminamos con ruido propio de la señal y, luego de pasarla por un ADC, incluimos ruido de cuantización al la señal de salida.
<p style='text-align: left;'>A travez de los graficos pudimos visualizar los resultados discutidos y entender como es el proceso aditivo de una señal de ruido sobre una señal de entrada. Para simplificar el análisis, la señal utilizada es una señal deterministica, de manera de conocer su comportamiento en el tiempo.
<p style='text-align: left;'>A su vez, analizamos el concepto de piso de ruido de la señal, y variamos parámetros para entender mejor esta idea.
<p style='text-align: left;'>El ADC que simulamos trabaja con $fs = 1000 Hz$ y un rango de señal de $±Vf = 2 Volts$. Utilizamos un primer modelo de señal con B = 4 bits y $Kn = 1$

En el primer grafico se puede ver la señal muestreada en un segundo. Se ve claramente su naturaleza senoidal. Superpuesto esta la señal con ruido analógico (azul) y la señal cuantizada (verde). Si bien ambas señales siguen a la onda original logicamente, se puede ver en especial con la señal cuantizada como toma ciertos valores por encima y por debajo de la onda original.
<p style='text-align: left;'>En el segundo grafico esta la representación en frecuencia de las ondas en relaciona su densidad de potencia.
<p style='text-align: left;'>El ultimo grafico, muestra como es la distribución del ruido de cuantización para distintos valores de voltajes.

Veremos ahora que sucede modificando valores de $Kn$ y la cantidad de Bits del ADC:

#### Caso 1: B = 4; Kn = 1/10

#### Caso 2: B = 8; Kn = 1

#### Caso 3: B = 16; Kn = 10

A fin de poder comparar mejor las 3 situaciones, pondremos juntos los graficos iguales para cada caso

En la medida que aumentamos el $Kn$ uno esperaría que la señal de ruido se distorcione aun mas sobre la orginal y los picos por encima y por debajo de la señal original serian todavia mas pronunciados. Sin embargo, el efecto de aumentar la cantidad de bits parece ser mas poderoso en cuanto a la mejora de la resolución del ADS. Lo cual es lógico, al aumentar la cantidad de bits, equivale a aumentar la cantidad de celdas en el grafico de la señal. Por lo que la diferencia entre el valor superior en inferior sobre el paso de la señal (ruido de cuantización), es menor. Es decir, hay mejor resolución en la representación cuantizada de la señal.

En el grafico de las densidades de potencia sin embargo, se puede ver como el aumento de la potencia de ruido infiere en un acercamiento hacia el piso de ruido analógico.

Los primeros gráficos muestran una relación entre la cantidad de bits y la resolución del ADS. Aumentar la cantidad de bits mejora significativamente la resolución del ADS porque disminuye el ruido de cuantización. Incluso al modificar el factor de potencia, la representación de la onda luego del ADS es mucho mejor porque los margenes de ruido son mas chicos. Aumentar la cantidad de bits, equivale a aumentar el numero de boxes en la grilla de representación entre $±Vf$.
