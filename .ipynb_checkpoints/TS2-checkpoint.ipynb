## Respuesta en frecuencia de filtros pasivos RLC

<p style='text-align: left;'>El objetivo de este trabajo fue analizar distintos tipos de filtros, a fin de poder familiarizarnos con su funcionamiento. 
<p style='text-align: left;'>Determinamos la función transferencia para cada uno y graficamos su respuesta en modulo y fase.
<p style='text-align: left;'>Analizamos intuitivamente el comportamiento del circuito en extremos de frecuencia para determinar el comportamiento de la red y definir de que tipo de filtro se trata.
De esta forma obtenemos informacion previa al analisis para saber que características debería presentar el circuito y comparar los resultado obtenidos.

### Circuitos:

En un primer analisis, consideramos las impedancias:

<p style='text-align: center;'>$Zc = 1/sC$ 
<p style='text-align: center;'>$Zl = sL$ 
<p style='text-align: center;'>$Zr = R$

<p style='text-align: left;'>Teniendo en cuenta esto, analizamos cada circuito:
<p style='text-align: left;'>Para el circuito A, cuando las frecuencias tienden a 0, la impedancia del capacitor tiende a infinito, por lo que no hay paso de corriente (se comporta como un circuito abierto). En el extremo opuesto, cuando las frecuencias tienden al infinito, la impedancia de la bobina tiende a infinito, por lo que no hay un paso de corriente tampoco en ese caso. Es decir que el caso intermedio es donde resta analizar, y estariamos hablando entonces de un <ins>Filtro pasa bandas</ins>.
<p style='text-align: left;'>Para el circuito B, cuando las frecuencias tienden a 0, sucede lo mismo que en el caso anterior, por lo que se filtrarán las bajas frecuencias. En el caso de las frecuencias en el infinito, la bobina tiene impedancia infinita y el circuito estaría reducido a un filtro RC convencional. Entonces estaríamos hablando de un <ins>Filtro pasa altos</ins>, ya que no permite el paso de frecuencias bajas.
<p style='text-align: left;'>Segun lo leido en bibliografía, la frecuencia de corte será donde la ganancia sera -3dB. En el caso del filtro pasa banda, habra dos frecuencias (*f1* y *f2*) que determinarán el ancho de banda y que ceran "equidistantes" a la frecuencia de corte.
<p style='text-align: left;'>En el filtro pasa altos, la frecuencia de corte determinará a partir de donde comienza el ancho de banda, que, en principio, sería infinita hacia las frecuencias en el infinito.

### Calcular la función transferencia de cada circuito

Para el filtro A:

### <p style='text-align: center;'>$H(s) = \frac{s(\omega0/Q)}{s^2+s(\omega0/Q)+\omega0^2}$

Para el filtro B:

### <p style='text-align: center;'>$H(s) = \frac{s^2}{s^2+s(\omega0/Q)+\omega0^2}$

### Respuesta de modulo y fase

Para calcular la respuesta en modulo de la frecuencia de ambos filtros, debemos hacer la conversión, en la función transferencia, de  "s" a "jw" y aplicarle el modulo.

#### Filtro A:

### <p style='text-align: center;'>$|H(s)| = \frac{\omega*\omega0/Q}{\sqrt{(\omega0^2-\omega^2)^2+(\omega*\omega0/Q)^2}}$

#### Graficos

El problema original no define parámetros para la frecuencia de corte. Por comodidad, utilizo $\omega0 = 1$ y $Q = 1$. De esta manera, los polos deberian estar dentro de la circunferencia unitaria en el mapa de polos y facilita el análisis.
Utilicé un ejemplo de código para graficar el mapa de ceros, la respuesta en módulo y en fase para aguilizar el proceso de elaboración del informe. Si bien es importante poder hacer mi propio codigo, entiendo que es aun mas importante la comprensión de los resultados. 

En el mapa se puede visualizar un cero en el origen, lo que nos dice que la ganancia del filtro es nuela cuando la frecuencia es 0.
Por otro lado, hay dos polos complejos marcados con cruces en la circunferencia unitaria (como era de esperarse). Los polos complejos permiten hacer una analisis rapido e intuitivo sobre la fase de la respuesta. Si pensamos por ejemplo en un movimiento hacia el infinito en el eje imaginario, los angulos que se forman en los polos tienden a $\Pi/2$. Al estar a cada lado del eje real, es de esperarse que la fase se encuentre entre $\Pi/2$ y $-\Pi/2$.
<p style='text-align: left;'>Por otra parte, la distancia que hay entre los polos y el eje imaginario, esta determinada por $Q$. De manera que si $Q$ es muy chico, los polos tienden a juntarse sobre el eje real.

El grafico de magnitud, es una campana simétrica en 1 rad/s. Esto quiere decir que a esa frecuencia (previamente seteada cuando dimos el valor de $\omega0$) es el minimo de atenuación del filtro. Por lo tanto, este filtro atenúa a bajas y altas frecuencias y permite el paso de frecuencias en una banda media.
<p style='text-align: left;'>Por otro lado, este filtro pasa bandas tiene una atenuación de 20 dB/dec, esto quiere decir que si por ejemplo tomamos el punto a 10 rad/s y a 100 rad/seg, hay una disminución de 20 db. Esta atenuación puede suceder aún de forma mas abrupta, dependiendo de $Q$.

Como deciamos antes, la respuesta en fase se mantiene en un rango entre $\Pi/2$ y $-\Pi/2$. El aporte que realizan los polos complejos, esta relacionado con este comportamiento.
<p style='text-align: left;'>Si analizamos el mapa de ceros a frecuencias bajas, el aporte que realizan los polos es de $\Pi/2$ que es donde comienza el grafico. En la medida que nos acercamos a la frecuencia de 1 rad/s (banda de minima atenuación en el modulo), la fase resultante se hace 0, ya que los aportes de ambos polos se cancelan.
<p style='text-align: left;'>Luego al movernos hacia frecuencias mas altas, el aporte resultante es de $-\Pi/2$ y se corresponde a la parte final del grafico de fase.

#### Filtro B:

### <p style='text-align: center;'>$|H(s)| = \frac{\omega^2}{\sqrt{(\omega0^2-\omega^2)^2+(\omega*\omega0/Q)^2}}$

Realizamos el mismo análisis que para el filtro A, el mapa de polos es similar, por lo que no lo incluyo. Pero se desprende el mismo analisis en relaciona a la ubicación de los polos y los ceros. Utilizamos los mismos valores de $Q$ y $\omega0$ para simplificar el análisis.
Tambien tiene un cero en el eje de cordenadas, esto sustenta el hecho de que a bajas frecuencias hay una ganancia nula del filtro y es consistente con tratarse de un filtro pasa altos. Luego el aporte de los polos conjugados, detrminarán el comportamiento del filtro.

El grafico de modulo muestra como a bajas frecuencias (antes de la frecuencia de corte $\omega0$) el filtro atenúa la señal y luego se estabiliza en 0. Es decir que a partir de $\omega0$, el filtro permite el paso de las frecuencias sin atenuarlas.
<p style='text-align: left;'>Se puede ver la atenuación de -20dB/dec en el margen de bajas frecuencias tomando dos puntos consecutivos.


La respuesta de fase coincide con un filtro pasa altos, donde la fase es positiva a bajas frecuencias, se mantiene positiva al pasar por rangos cercanos a la frecuencia de corte $\omega0$, y luego se hace negativa para frecuencias altas.

## Apendice

### Calculo de Funciones transferencia:

* Filtro A:

<p style='text-align: left;'>$V0 = R$
<p style='text-align: left;'>$Vi = \frac{1}{sC} + sL + R$
<p style='text-align: left;'>$H(s) = \frac{R}{\frac{1}{sC} + sL + R}$
<p style='text-align: left;'>$H(s) = \frac{sRC}{s^2LC + sCR + 1}$
<p style='text-align: left;'>Si $\omega0^2 = \frac{1}{LC}$ y $\frac{\omega0}{Q} = \frac{R}{L}$
<p style='text-align: left;'>$H(s) = \frac{s*(R/L)}{s^2 + (s*R)/L + \omega0^2}$
<p style='text-align: left;'>$H(s) = \frac{s(\omega0/Q)}{s^2+s(\omega0/Q)+\omega0^2}$

* Filtro B:

<p style='text-align: left;'>$V0 = \frac{sRL}{sL+R}$
<p style='text-align: left;'>$Vi = \frac{sRL}{sL+R}+\frac{1}{sC}$
<p style='text-align: left;'>$Vi = \frac{sL+R+s^2LRC}{sC*(sL+R)}$
<p style='text-align: left;'>$H(s) = \frac{\frac{sRL}{sL+R}}{\frac{sL+R+s^2LRC}{sC*(sL+R)}}$
<p style='text-align: left;'>Si $\omega0^2 = \frac{1}{LC}$ y $\frac{\omega0}{Q} = \frac{1}{RC}$
<p style='text-align: left;'>$H(s) = \frac{s^2}{s^2 + s/(RC) + 1/(LC)}$
<p style='text-align: left;'>$H(s) = \frac{s^2}{s^2+s(\omega0/Q)+\omega0^2}$
