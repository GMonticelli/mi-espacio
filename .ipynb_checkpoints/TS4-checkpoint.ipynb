## Primeras nociones de estimación espectral

En este trabajo modelizamos una señal senoidal contaminada con ruido propio de la señal y calculamos los estimadores de la amplitud y de la frecuencia.
<p style='text-align: left;'>A travez del grafico de la densidad de potencia, pudimos estimar la frecuencia a la cual se visualiza la potencia máxima. Esta es la frecuencia que usamos como referencia para aplicar distintas ventanas.

### Espectro densidad de potencia:

Como se visualiza en el gráfico, la frecuencia de 250 Hz es un pico de potencia. En esa frecuencia aplicamos la ventana rectangular (por defecto), la ventana Blackmanharris y la ventana Flattop.

El gráfico representa el calculo del estimador $a1$, es decir, la cantidad de veces (eje Y) que cada valor de amplitud (eje X) aparece para una frecuencia de 250 Hz en cada una de las ventanas mencionadas anteriormente.
<p style='text-align: left;'>Luego realizaremos el cálculo de la varianza y sesgo para saber cual de las tres ventanas funciona mejor para el estimador.

Para calcular el segundo estimador ($\Omega$1), calculamos la potencia de la transformada con las distintas ventanas. Buscamos el máximo de ese valor y lo multiplicamos por el espectro para obtener el valor maximo de la frecuencia.

### Conclusión

En el caso del estimador $\Omega$1, si bien las diferencias son muy chicas, la ventana recangular tiene menor sesgo y varianza en relación a las otras dos. Por lo que resulta mejor para ese estimador.
<p style='text-align: left;'> Para el estimador $a1$, la ventana rectangular tiene menosr sesgo en ambos casos de SNR, sin embargo la ventana flattop tiene menos varianza. Siendo que el sesgo puede ser mas facil corregirlo, la diferencia de varianza tiene mayor peso. Y por eso en este caso sería mejor elegir una ventana flattop para ese estimador.
